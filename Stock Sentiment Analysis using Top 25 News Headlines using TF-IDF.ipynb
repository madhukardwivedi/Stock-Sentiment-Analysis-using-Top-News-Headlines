## Stock Sentiment Analysis using Top 25 News Headlines using TF-IDF

    The data set in consideration is a combination of the world news and stock price shifts.
    Data ranges from 2008 to 2016 and the data from 2000 to 2008 was scrapped from Yahoo finance.
    There are 25 columns of top news headlines for each day in the data frame.
    Class 1- the stock price increased.
    Class 0- the stock price stayed the same or decreased.

```python
## import all the importnat libraries 
import pandas as pd
import numpy as np
import nltk
```

```python
data = pd.read_csv('D:/ML_learning/Stock_Sentiment_Analysis_NLP/Data.csv')
data.head()
```

```python
## Lest convert the data into train and test
train = data[data['Date'] < '20150101']
test = data[data['Date'] > '20141231']
```

```python
test.shape
```

```python
train.head()
```

```python
### now apply data cleaning
data = train.iloc[:, 2:27]

# Now replace all the other characters other than a-z and A-Z
data.replace("[^a-zA-Z]"," ",regex=True, inplace=True)

## Now change the names of columns for better understanding

list1 = (i for i in range(25))
new_index = [str(i) for i in list1]
data.columns = new_index
data.head(2)
```

```python
## Lower the headlines

for index in new_index:
    data[index] = data[index].str.lower()
```

```python
data.head(1)
```

```python
## Now we have headlines in different columns. We have to convert these headlines into one paragraph

headlines = []
for i in range(0, len(data.index)):
    merged = ' '.join(str(x) for x in data.iloc[i, 0:25])
    headlines.append(merged)
```

```python
headlines[0]
```

```python
## Now create the Machine Learning model

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.ensemble import RandomForestClassifier 
```

```python
## Implement Bag of Words which will convert all the words to vectors

# implement TF-IDF
TF_IDF = TfidfVectorizer(ngram_range=(2,2))
traindataset=TF_IDF.fit_transform(headlines)
```

```python
# implement RandomForest Classifier
randomclassifier=RandomForestClassifier(n_estimators=200,criterion='entropy')
randomclassifier.fit(traindataset,train['Label'])
```

```python
## Perform data cleaning for test data as well. 

clean_test = test.iloc[:,2:27]
clean_test.replace("[^a-zA-Z]", ' ', regex = True, inplace = True)

## change the name of the columns
list2= [i for i in range(25)]
new_Index2=[str(i) for i in list2]
clean_test.columns= new_Index2
clean_test.head(5)

## Lower the sentences
for index in new_Index2:
    clean_test[index]=clean_test[index].str.lower()

## Combine all the headlines from different columns into one
test_transform= []
for row in range(0,len(clean_test.index)):
    test_transform.append(' '.join(str(x) for x in clean_test.iloc[row,0:25]))

test_transform[0]
```

```python
test_dataset = TF_IDF.transform(test_transform)

predictions = randomclassifier.predict(test_dataset)
```

```python
## Import library to check accuracy
from sklearn.metrics import classification_report,confusion_matrix,accuracy_score

matrix = confusion_matrix(test['Label'],predictions)
print(matrix)
score = accuracy_score(test['Label'],predictions)
print(score)
report = classification_report(test['Label'],predictions)
print(report)
```

```python
## Using Naive Bayes
from sklearn.naive_bayes import MultinomialNB

nb = MultinomialNB()
nb.fit(traindataset,train['Label'])
```

```python
predictions_nb = nb.predict(test_dataset)
```

```python
matrix = confusion_matrix(test['Label'],predictions_nb)
print(matrix)
score = accuracy_score(test['Label'],predictions_nb)
print(score)
report = classification_report(test['Label'],predictions_nb)
print(report)
```
